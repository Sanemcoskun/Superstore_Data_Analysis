```python
df = data.drop(columns=["Row ID","Order ID","Customer ID","Product ID"]) # You can use axis=1
```

```python
data.isnull().sum()
```

```python
df[df["Postal Code"].isnull()] #Null values
```

```python
df["Postal Code"] = df["Postal Code"].fillna(5402)
```

```python
df.isnull().sum()
```

```python
df[df["City"]=="Burlington"] # Fillna function control
```

```python
# Month column
df["Order_date"] = pd.to_datetime(df["Order Date"],format= "%d/%m/%Y")
df["Order_Month"] = df["Order Date"].str[3:5].astype(int)
df.head()
```

```python
# Month to name
month_name = {1: 'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'}
month_name[1]
    
```

```python
df["Order_Month"] = [month_name[i] for i in df["Order_Month"]]
df.head()
```

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"],format= "%d/%m/%Y")
```

```python
df
```

```python
df["Year"] = df["Order_date"].dt.year
```

```python
df
```

```python
df["Year"].value_counts()
```

```python
df["Order_Month"].value_counts()
```

```python
# Which month has more sales?
months_df = df.groupby("Order_Month").sum(numeric_only=True).sort_values(by="Sales",ascending=False)
months_df
```

```python
months_df = months_df[["Sales"]] # just selected Sales
# if you use one []. t returns Series Else returns DataFrame
months_df
```

```python
months_df.reset_index(inplace=True)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=months_df,x=months_df.index,y= months_df["Sales"],palette="Set2")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
year_df = df.groupby("Year").sum(numeric_only=True)
year_df = year_df[["Sales"]]
year_df
```

```python
year_df.reset_index(inplace=True)
year_df
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=year_df,x=year_df["Year"],y= year_df["Sales"],palette="Set2")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.title("Sales Over Years")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
customer = df.groupby("Customer Name").sum(numeric_only=True).sort_values(by="Sales",ascending=False) # 
customer.drop(["Postal Code","Year"],axis=1,inplace=True) # original df is permanently changed
customer.reset_index(inplace=True)
customer
```

```python
top_customers = customer.head(10)

plt.figure(figsize=(8,8))
plt.pie(top_customers["Sales"], labels=top_customers["Customer Name"], autopct="%1.1f%%", startangle=140)
plt.title("Top 10 Customers by Sales (Percentage)")
plt.tight_layout()
plt.show()
```

```python
category = df.groupby("Category").sum(numeric_only=True)
category.drop(["Postal Code","Year"],axis=1,inplace=True)
category.reset_index(inplace=True)
category
```

```python
explode = [0] * len(category) # Initialize all slices with no offset
max_index = category["Sales"].idxmax() # Find index of category with the highest sales
explode[category.index.get_loc(max_index)] = 0.1 # Pull out the biggest slice slightly
```

```python

plt.figure(figsize=(8,8))
plt.pie(category["Sales"],labels=category["Category"],autopct="%1.1f%%",startangle=140,explode=explode)
# explode (0.1,0,0)
plt.title("Categories")
plt.tight_layout()
plt.show()
```
